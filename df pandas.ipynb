```python
df = pd.merge(left=df, right=df_cities, how='inner',left_on='city_id', right_on='id') #Une ambos df en base a esas claves
```

```python
cat_0_bars = df['category_0'] == 'Bars'
cat_1_bars = df['category_1'] == 'Bars'

carnegie = df['city'] == 'Carnegie'
df[(cat_0_bars | cat_1_bars) & carnegie]
```

```python
cat_0 = df['category_0'].isin(['Bars', 'Restaurants'])
cat_1 = df['category_1'].isin(['Bars', 'Restaurants'])
carnegie = df['city'] == 'Carnegie'
df[(cat_0 | cat_1) & carnegie]
```

```python
lv = df['city'] == 'Las Vegas'
cat_0_db = df['category_0'] == 'Dive Bar'
cat_1_db = df['category_1'] == 'Dive Bar'
divebars_lv = df[lv & (cat_0_db | cat_1_db)]
```

```python
len(divebars_lv)
divebars_lv.head()
```

```python
stars = divebars_lv['stars'] >= 4.0
divebars_lv_4star = divebars_lv[stars]
divebars_lv_4star.head(53)
```
